```python
import numpy as np
import skimage.io
import matplotlib.pyplot as plt
```

```python
clown = skimage.io.imread('http://www.imagemagick.org/Usage/fourier/clown_orig.jpg')
twigs = skimage.io.imread('http://www.imagemagick.org/Usage/fourier/twigs.jpg', as_gray = True)
```

## Fourier transformation - picture "twigs"

```python
# Fourier transformation 1
ft = np.fft.fft2(twigs) # 2D Fourier Transform (FT)
ftshift = np.fft.fftshift(ft) # Change quadrants of the FT (1<-->3, 2<-->4)
spectrum = 20*np.log(np.abs(ftshift)) # Spectrum of FT
plt.imshow(spectrum, cmap = 'gray')
```

```python
# Creating mask
img_shape = twigs.shape
mask = np.ones([img_shape[0], img_shape[1]])
mask[126:130, 0:85] = 0
mask[172:255, 126:130] = 0

mask[126:130, 170:255] = 0
mask[0:85, 127:131] = 0
```

```python
# Product
product = (mask * ftshift)
restored_twigs = np.abs(np.fft.ifft2(product))
```

```python
# Fourier transformation 2
ft = np.fft.fft2(restored_twigs) # 2D Fourier Transform (FT)
ftshift = np.fft.fftshift(ft) # Change quadrants of the FT (1<-->3, 2<-->4)
spectrum_restored_im = 20*np.log(np.abs(ftshift)) # Spectrum of FT
plt.imshow(spectrum_restored_im, cmap = 'gray')
```

```python
# Plotting
f, axarr = plt.subplots(1,4, figsize=(18,6))
axarr[0].imshow(twigs, cmap="gray")
axarr[0].set_title('Original picture')
axarr[1].imshow(spectrum, cmap = "gray")
axarr[1].set_title('Spectrum of the original image')
axarr[2].imshow(restored_twigs, cmap = "gray")
axarr[2].set_title('Restored picture')
axarr[3].imshow(spectrum_restored_im, cmap = "gray")
axarr[3].set_title('Spectrum of the restored picture')
plt.show()
```

## Fourier transformation - picture "clown"

```python
# Fourier transform 1
ft_clown = np.fft.fft2(clown) # 2D Fourier Transform (FT)
ftshift_clown = np.fft.fftshift(ft_clown) # Change quadrants of the FT (1<-->3, 2<-->4)
spectrum_clown = 20*np.log(np.abs(ftshift_clown)) # Spectrum of FT

# Plotting
f, axarr = plt.subplots(1,2, figsize=(18,6))
axarr[0].imshow(clown, cmap="gray")
axarr[0].set_title('Original picture')
axarr[1].imshow(spectrum_clown, cmap = "gray")
axarr[1].set_title('Spectrum of the original image')
plt.show()
```

```python
# Creating mask (k odstraneni "hvezdicek ze spektra")
img_shape = clown.shape
mask = np.ones([img_shape[0], img_shape[1]])

mask[86:89, 10:40] = 0
mask[70:100, 23:25] = 0

mask[52:53, 40:60] = 0
mask[48:58, 45:46] = 0

mask[75:80, 80:100] = 0
mask[65:85, 85:90] = 0

mask[40:45, 95:118] = 0
mask[25:50, 100:110] = 0
```

```python
# Product - clown
product = (mask * ftshift_clown)
restored_clown = np.abs(np.fft.ifft2(product))
```

```python
ft_res_im = np.fft.fft2(restored_clown) # 2D Fourier Transform (FT)
ftshift_res_im = np.fft.fftshift(ft_res_im) # Change quadrants of the FT (1<-->3, 2<-->4)
spectrum_restored_clown = 20*np.log(np.abs(ftshift_res_im)) # Spectrum of FT
#plt.imshow(spectrum_restored_clown, cmap = 'gray')
```

```python
# Plotting
f, axarr = plt.subplots(1,4, figsize=(18,6))
axarr[0].imshow(clown, cmap="gray")
axarr[0].set_title('Original picture')
axarr[1].imshow(spectrum_clown, cmap = "gray")
axarr[1].set_title('Spectrum of the original image')
axarr[2].imshow(restored_clown, cmap = "gray")
axarr[2].set_title('Restored picture')
axarr[3].imshow(spectrum_restored_clown, cmap = "gray")
axarr[3].set_title('Spectrum of the restored picture')
plt.show()
```
